```python
from pandas_datareader import data as pdr
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
#stocks = ['BTC-USD','ETH-USD','USDT-USD','BNB-USD','ADA-USD','XRP-USD','SOL-USD','DOGE-USD','DOT-USD','UNI-USD','BTT2-USD','TRX-USD','NEO-USD', 'XVG-USD','XMR-USD']
BTC_Coin = ['BTC-USD']
ETH_Coin = ['ETH-USD']
BNB_Coin = ['BNB-USD']
```

```python
def get_data(tickers, start_date, end_date):
        
    # download daily price data for each of the stocks in the portfolio
    df = pdr.get_data_yahoo(stocks, start=start_date, end = end_date)['Adj Close']
    df.sort_index(inplace = True)
    
    return df
```

```python
yf.pdr_override() # <== that's all it takes :-)
# download dataframe
BTC = pdr.get_data_yahoo(BTC_Coin, start="2021-01-01", end="2022-07-25")
ETH = pdr.get_data_yahoo(ETH_Coin, start="2021-01-01", end="2022-07-25")
BNB = pdr.get_data_yahoo(BNB_Coin, start="2021-01-01", end="2022-07-25")
```
